```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("googleplaystore.csv")

# Remove linhas duplicadas
df.drop_duplicates(inplace=True)

# Converter a coluna 'Category' para string
df['Category'] = df['Category'].astype(str)

# Lista das categorias importantes
important_categories = ["FAMILY", "GAME", "TOOLS", "MEDICAL", "BUSINESS", "PRODUCTIVITY", "PERSONALIZATION", "COMMUNICATION", "SPORTS"]

# Funo para categorizar como 'OUTROS' se no estiver nas categorias importantes
def categorize_category(category):
    if category.upper() in [cat.upper() for cat in important_categories]:
        return category
    else:
        return "OUTROS"

# Aplicar a funo para criar uma nova coluna de categorias ajustadas
df['Adjusted_Category'] = df['Category'].apply(categorize_category)

# Contagem da frequncia de cada categoria ajustada
category_counts = df['Adjusted_Category'].value_counts()

# Criao do grfico de pizza
plt.figure(figsize=(10, 7))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.2f%%', startangle=140)
plt.title('Distribuio das Categorias de Apps')
plt.axis('equal')  # Assegura que o grfico seja desenhado como um crculo.
plt.show()
```
